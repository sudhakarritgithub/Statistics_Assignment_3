```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at point x.

    Parameters:
    - x (float): The point at which to evaluate the PDF.
    - mean (float): Mean of the normal distribution.
    - std_dev (float): Standard deviation of the normal distribution.

    Returns:
    - float: PDF value at point x.
    """
    exponent = -((x - mean)**2) / (2 * std_dev**2)
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

# Example usage:
mean = 0
std_dev = 1
x = 1.5
pdf_at_x = normal_pdf(x, mean, std_dev)
print(f"PDF at x = {x}: {pdf_at_x}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 20   # Number of trials
p = 0.4  # Probability of success

# Generate random sample of size 1000 from Binomial distribution
sample = np.random.binomial(n, p, size=1000)

# Plotting histogram
plt.figure(figsize=(8, 6))
plt.hist(sample, bins=np.arange(0, n+2)-0.5, density=True, edgecolor='black')  # bins for discrete values
plt.xticks(np.arange(0, n+1))
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.title(f'Binomial Distribution: n={n}, p={p}')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import math

def poisson_cdf(k, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution at point k.

    Parameters:
    - k (int): The point at which to evaluate the CDF.
    - mean (float): Mean (lambda) of the Poisson distribution.

    Returns:
    - float: CDF value at point k.
    """
    cdf = 0.0
    for i in range(k + 1):
        cdf += math.exp(-mean) * (mean**i) / math.factorial(i)
    return cdf

# Example usage:
mean = 2.5
k = 3
cdf_at_k = poisson_cdf(k, mean)
print(f"CDF at k = {k}: {cdf_at_k}")

```

```python
import numpy as np

# Parameters
mean = 5  # Mean (lambda) of the Poisson distribution
sample_size = 1000  # Size of the random sample

# Generate random sample from Poisson distribution
sample = np.random.poisson(mean, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```
